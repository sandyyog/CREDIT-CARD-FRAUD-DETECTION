```python
import pandas as pd
import numpy as np
credit_data=pd.read_csv('credit card datas.txt')
credit_data
```

```python
### checking the no of missing values
credit_data.isnull().sum()
```

```python
### distribution of legit transcations & fraudulent transcation
credit_data['Class'].value_counts()
```

```python
### This dataset is highly unbalanced
0----->Normal transcation
1------>Fradulent transcation
```

```python
### separating the data for analysis
legit=credit_data[credit_data.Class==0]
fraud=credit_data[credit_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
### statistical measures of the data 
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
### compare the values for both transcation
credit_data.groupby('Class').mean()
```

```python
### build a sample dataset constaining similar distribution of normal transcation and fradulent transcation
legit_sample=legit.sample(n=2)
```

```python
## concatenating two dataframes
new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
###splitting the data into features&targets
x=new_dataset.drop(columns='Class',axis=1)
print(x)
```

```python
y=new_dataset['Class']
print(y)
```

```python
### train & test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape,x_test.shape)
```

```python
print(y_train.shape,y_test.shape)
```

```python
### model_traing
from sklearn.linear_model import LogisticRegression
obj=LogisticRegression()
obj.fit(x_train,y_train)
```

```python
### accuracy
from sklearn.metrics import accuracy_score
x_train_prediction=obj.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print(training_data_accuracy)
```

```python
### accuracy on testing data
x_test_prediction=obj.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print(test_data_accuracy)
```
